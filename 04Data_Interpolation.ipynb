DATA INTERPOLATION



1. Linear Interpolation

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# Interplote the data using linear interploation
x_new = np.linspace(1,5,10) ## create new x values
y_interp = np.interp(x_new,x,y) ## interpolate y values
print(y_interp)
```

```python
plt.scatter(x_new,y_interp)
```

## Cubic Interpolation with Scipy

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])
```

```python
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
## create a cubic interpolation function
f = interp1d(x,y,kind = 'linear')
```

```python
## interpolate the data
x_new = np.linspace(1,5,10)
y_interp=f(x_new)
print(y_interp)
```

```python
plt.scatter(x_new,y_interp)
```

## Polinomial Interpolation

```python
import numpy as np

# create some sample data
x = np.array([1,2,3,4,5])
y = np.array([1,2,9,16,15])
```

```python
# interpolate the data using polynomial interpolation
p = np.polyfit(x,y,4) # fit a 2nd degree polynomial to the data
```

```python
x_new = np.linspace(1,5,10) # create new x values 
y_interp = np.polyval(p,x_new) # interpolate y values
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```
